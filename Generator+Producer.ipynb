Imports

```python
from confluent_kafka import Producer
import requests
import json
import time
```

Functions

```python
def get_exchange_list():
    url = "https://api.coingecko.com/api/v3/exchanges"
    response = requests.get(url)
    return response.json()

def get_trading_volume(exchange_id, days=1):
    url = f"https://api.coingecko.com/api/v3/exchanges/{exchange_id}/volume_chart?days={days}"
    response = requests.get(url)
    return response.json()

def exchange_volume_generator():
    exchanges = get_exchange_list()
    for exchange in exchanges:
        volume_data = get_trading_volume(exchange['id'], days=1)
        message = json.dumps({'name': exchange['name'], 'volume': volume_data}).encode('utf-8')
        print(f"Sending data for Exchange ID: {exchange['id']}, Volume: {volume_data}")
        yield message
        print('Waiting 12 seconds before sending new request to API')
        time.sleep(12)       
```

Producer

```python
producer_conf = {'bootstrap.servers': 'localhost:9092'}
p = Producer(producer_conf)

topic = 'exchange_volume_topic'

for message in exchange_volume_generator():
    p.produce(topic, message)
    time.sleep(1)  

p.flush()
```
